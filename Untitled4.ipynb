```python
import pandas as pd

# Load your dataset
df = pd.read_csv("Discretized_Dataset_for_Bayesian_Network.csv")

# Identify condition columns
condition_cols = [col for col in df.columns if col.startswith("has_")]

# Define rule-based function to classify condition context
def classify_condition(row):
    cardiac = any(row.get(c, 0) == 1 for c in condition_cols if "cardiac" in c or "heart" in c or "hypertension" in c)
    respiratory = any(row.get(c, 0) == 1 for c in condition_cols if "asthma" in c or "respiratory" in c or "breath" in c)
    diabetes = any(row.get(c, 0) == 1 for c in condition_cols if "diabetes" in c or "insulin" in c)
    mental = any(row.get(c, 0) == 1 for c in condition_cols if "stress" in c or "depression" in c or "anxiety" in c)

    types = sum([cardiac, respiratory, diabetes, mental])
    if types > 1:
        return "mixed"
    elif cardiac:
        return "cardiac_related"
    elif respiratory:
        return "respiratory"
    elif diabetes:
        return "diabetes_related"
    elif mental:
        return "mental_health"
    else:
        return "none"

# Apply function across all rows
df['condition_context'] = df.apply(classify_condition, axis=1)

# Save new file
df.to_csv("Discretized_Dataset_with_Context.csv", index=False)
print("✅ New file saved: Discretized_Dataset_with_Context.csv")
```
