```python
import os
import pandas as pd

# Load the cleaned dataset
df = pd.read_csv("Cleaned_Dataset_with_Encoded_Triage.csv")
df = df.astype(str).fillna("N/A")

# Output directory
output_dir = "documents_from_cleaned_csv"
os.makedirs(output_dir, exist_ok=True)

# Write each row to a text file
for idx, row in df.iterrows():
    filename = f"document_{idx}.txt"
    file_path = os.path.join(output_dir, filename)
    
    with open(file_path, "w", encoding="utf-8") as f:
        for col in df.columns:
            f.write(f"{col}: {row[col]}\n")

print(f"✅ Saved {len(df)} documents in '{output_dir}'")
```
