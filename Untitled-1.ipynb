# Bayesian Network Inference and Analysis

This notebook demonstrates how to perform inference queries on a pre-trained Bayesian Network model saved in `learned_bn_model.json`. It also shows how to analyze the impact of specific features on the target variable (`triage_level`) using inference.

**Steps:**
1.  **Setup:** Load necessary libraries and the pre-trained Bayesian Network model. Load the data subset required for baseline calculations.
2.  **Initialize Inference Engine:** Create an inference object using `VariableElimination`.
3.  **Define Evidence Scenarios:** Specify different sets of observed evidence.
4.  **Prepare Evidence Function:** Define a helper function to format evidence correctly for queries.
5.  **Execute Inference Queries:** Run queries for the defined scenarios.
6.  **Analyze Feature Impact:** Perform sensitivity analysis by changing individual features and observing the impact on the target variable's probability distribution.

## 1. Setup: Load Libraries, Model, and Data
Import necessary libraries. Load the Bayesian Network structure and parameters (CPDs) from the JSON file. Reconstruct the `pgmpy` model object. Load the relevant data subset to calculate baseline evidence (mode of parent variables).

```python
import pandas as pd
import numpy as np
import json
import networkx as nx # Needed for potential cycle checks if modifying structure
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.factors.discrete import TabularCPD
from pgmpy.inference import VariableElimination
from sklearn.preprocessing import LabelEncoder # For potential data loading/re-encoding if needed

# --- Load Model from JSON ---
model_file = "learned_bn_model.json"
print(f"--- Loading Model from {model_file} ---")

try:
    with open(model_file, "r") as f:
        model_data = json.load(f)

    # Create an empty Bayesian Network
    model = DiscreteBayesianNetwork()

    # Add nodes and edges
    nodes = model_data["nodes"]
    edges = [(str(u), str(v)) for u, v in model_data["edges"]]
    model.add_nodes_from(nodes)
    model.add_edges_from(edges)
    print(f"Loaded {len(nodes)} nodes and {len(edges)} edges.")

    # Add CPDs
    state_names_map = {} # Reconstruct state names map
    for cpd_data in model_data["cpds"]:
        variable = str(cpd_data["variable"])
        evidence = [str(ev) for ev in cpd_data["evidence"]]
        values = np.array(cpd_data["values"])
        
        # Reconstruct state_names dictionary with integer states where appropriate
        cpd_state_names = {}
        raw_state_names = cpd_data["state_names"]
        for var, states_str in raw_state_names.items():
             # Attempt to convert states back to int if possible, otherwise keep as string
             try:
                 # Check if the variable was originally numeric/ordinal based on common names
                 # This is heuristic; ideally, type info would be saved in JSON
                 if var in ['Heart rate', 'Systolic Blood Pressure', 'Diastolic Blood Pressure',
                           'Respiratory rate', 'Body temperature',
                           'Pain severity - 0-10 verbal numeric rating [Score] - Reported', 'age',
                           'triage_level'] or all(s.isdigit() or (s.startswith('-') and s[1:].isdigit()) for s in states_str):
                     cpd_state_names[str(var)] = [int(s) for s in states_str]
                 else: # Keep as string if conversion seems inappropriate
                     cpd_state_names[str(var)] = [str(s) for s in states_str]
             except ValueError: # Fallback to string if int conversion fails
                 cpd_state_names[str(var)] = [str(s) for s in states_str]
        
        state_names_map.update(cpd_state_names) # Add/update global map

        # Get cardinalities from the length of state names lists
        variable_card = len(cpd_state_names[variable])
        evidence_card = [len(cpd_state_names[ev]) for ev in evidence]

        # Create TabularCPD
        cpd = TabularCPD(
            variable=variable,
            variable_card=variable_card,
            values=values,
            evidence=evidence,
            evidence_card=evidence_card,
            state_names=cpd_state_names # Use the reconstructed state names
        )
        model.add_cpds(cpd)
        # print(f"Added CPD for: {variable}") # Optional print

    print("CPDs loaded and added to the model.")

    # Check model validity
    model.check_model()
    print("✅ Model loaded and checked successfully.")

except FileNotFoundError:
    print(f"Error: Model file '{model_file}' not found. Cannot proceed.")
    model = None # Ensure model is None if loading fails
except Exception as e:
    print(f"Error loading or reconstructing model: {e}")
    model = None

# --- Load Data Subset for Baseline Calculation ---
# We need the original (or similarly preprocessed) data subset
# to calculate the mode for baseline evidence.
# Re-running the preprocessing steps from the original script:
print("\n--- Loading and Preprocessing Data for Baseline ---")
try:
    df_full = pd.read_csv("Discretized_Dataset_with_Context.csv")
    df_full = df_full.dropna() # Basic dropna

    # Manual ordinal encoding (copy from original script)
    ordinal_mapping = {'low': 0, 'medium': 1, 'high': 2}
    ordinal_columns = [
        'Heart rate', 'Systolic Blood Pressure', 'Diastolic Blood Pressure',
        'Respiratory rate', 'Body temperature',
        'Pain severity - 0-10 verbal numeric rating [Score] - Reported', 'age'
    ]
    for col in ordinal_columns:
        if col in df_full.columns:
            df_full[col] = df_full[col].map(lambda x: ordinal_mapping.get(x, x))

    # Label encode other categorical variables (copy from original script)
    label_encoders = {}
    for col in df_full.select_dtypes(include=['object']).columns:
        if col in df_full.columns:
            le = LabelEncoder()
            df_full[col] = le.fit_transform(df_full[col])
            label_encoders[col] = le # Store if needed, though not used directly here

    # Select columns used in the model (use nodes from loaded model)
    selected_columns = model.nodes()
    df_subset = df_full[[col for col in selected_columns if col in df_full.columns]].copy() # Use .copy()

    # Convert columns to appropriate integer types (copy from original script)
    for col in df_subset.columns:
        if not pd.api.types.is_numeric_dtype(df_subset[col]):
            try:
                df_subset[col] = pd.to_numeric(df_subset[col], errors='coerce').astype('Int64')
            except Exception as e:
                 print(f"Warning: Could not convert column {col} to numeric during reload. Error: {e}")
        elif pd.api.types.is_float_dtype(df_subset[col]):
             df_subset[col] = df_subset[col].astype(float).astype('Int64')
        elif pd.api.types.is_integer_dtype(df_subset[col]):
             df_subset[col] = df_subset[col].astype('Int64')

    # Drop rows with NaN potentially introduced by coercion
    df_subset = df_subset.dropna()
    print(f"Loaded and preprocessed data subset with {df_subset.shape[0]} rows and {df_subset.shape[1]} columns.")
    # print("Columns in df_subset:", df_subset.columns.tolist()) # Verify columns
    # print("Data types in df_subset:\n", df_subset.dtypes) # Verify types

except FileNotFoundError:
    print("Error: Data file 'Discretized_Dataset_with_Context.csv' not found. Cannot calculate baseline.")
    df_subset = None
except Exception as e:
    print(f"Error loading or preprocessing data: {e}")
    df_subset = None

# Check if model and data are ready
if model is None or df_subset is None:
    print("\n❌ Cannot proceed with inference due to errors in loading model or data.")
else:
    print("\n✅ Setup complete. Model and data subset are ready for inference.")
```

## 2. Initialize Inference Engine
Create an instance of the `VariableElimination` class using the loaded model.

```python
if model:
    infer = VariableElimination(model)
    print("Inference engine initialized.")
else:
    print("Model not loaded, cannot initialize inference engine.")
    infer = None # Ensure infer is None if model wasn't loaded
```

## 3. Define Evidence Scenarios
Define Python dictionaries representing different patient scenarios. The keys are variable names (matching the model's nodes), and the values are the observed states (encoded as integers).

*   **Low Evidence:** Represents a generally stable patient.
*   **Moderate Evidence:** Represents a patient with some concerning signs.
*   **High Evidence:** Represents a patient with critical signs.
*   **Cardiac Evidence:** Represents a patient with specific cardiac indicators.

*Note: The values used (0, 1, 2) correspond to the 'low', 'medium', 'high' or binary (0, 1) encoding used during model training.*

```python
# Define Evidence Scenarios (Ensure values match learned states/encoding)
# These are the raw scenarios before preparation.
# Values should be integers based on the encoding used (e.g., low=0, medium=1, high=2; False=0, True=1)

cardiac_evidence = {
    "Heart rate": 1, # medium
    "Systolic Blood Pressure": 1, # medium
    "Diastolic Blood Pressure": 1, # medium (Note: May not be a direct parent of triage_level)
    "Respiratory rate": 1, # medium
    "Pain severity - 0-10 verbal numeric rating [Score] - Reported": 0, # low
    "has_myocardial_infarction": 1, # True
    "has_diabetes": 0, # False
    "condition_context": 0, # Assuming 0 corresponds to 'cardiac_related' or similar after label encoding
    "has_cardiac_arrest": 0, # False
    "age": 1 # medium
}

low_evidence = {
    "Heart rate": 1, # medium
    "Systolic Blood Pressure": 1, # medium
    "Diastolic Blood Pressure": 1, # medium
    "Respiratory rate": 1, # medium
    "Pain severity - 0-10 verbal numeric rating [Score] - Reported": 0, # low
    "condition_context": 0, # Assuming 0 is a common/less severe context
    "has_diabetes": 0, # False
    "has_cardiac_arrest": 0, # False
    "has_myocardial_infarction": 0, # False
    "age": 1 # medium
}

moderate_evidence = {
    "Heart rate": 1, # medium
    "Systolic Blood Pressure": 1, # medium
    "Diastolic Blood Pressure": 1, # medium
    "Respiratory rate": 2, # high
    "Pain severity - 0-10 verbal numeric rating [Score] - Reported": 2, # high
    "condition_context": 2, # Assuming 2 is a more severe context
    "has_diabetes": 1, # True
    "has_cardiac_arrest": 1, # True
    "has_myocardial_infarction": 0, # False
    "age": 1 # medium
}

high_evidence = {
    "Heart rate": 2, # high
    "Systolic Blood Pressure": 0, # low
    "Diastolic Blood Pressure": 0, # low
    "Respiratory rate": 2, # high
    "Pain severity - 0-10 verbal numeric rating [Score] - Reported": 2, # high
    "condition_context": 0, # Assuming 0 is 'cardiac_related' or a relevant context
    "has_diabetes": 1, # True
    "has_cardiac_arrest": 1, # True
    "has_myocardial_infarction": 0, # False (Note: MI=0 but Cardiac Arrest=1)
    "age": 2 # high
}

print("Evidence scenarios defined:")
print("Cardiac Evidence:", cardiac_evidence)
print("Low Evidence:", low_evidence)
print("Moderate Evidence:", moderate_evidence)
print("High Evidence:", high_evidence)
```

## 4. Prepare Evidence Function
The `infer.query()` method requires evidence only for the parent nodes of the target variable(s). This function takes a raw evidence scenario, identifies the parents of `triage_level`, calculates baseline evidence (mode) for all parents, and then constructs the final evidence dictionary containing only parent variables, using the scenario's values where provided and baseline values otherwise. It also ensures correct data types.

```python
# --- Get Parents and Baseline ---
target_variable = 'triage_level'
parents = []
triage_parents_set = set()
baseline_evidence = {}

if model and df_subset is not None and target_variable in model.nodes():
    try:
        parents = model.get_parents(target_variable)
        print(f"\nParents of {target_variable}: {parents}")
        triage_parents_set = set(parents)

        all_parents_found_in_data = True
        for p in parents:
            if p in df_subset.columns:
                # Calculate mode, ensure it's a standard Python int
                mode_val = df_subset[p].mode()[0]
                if isinstance(mode_val, np.integer):
                    mode_val = int(mode_val)
                elif not isinstance(mode_val, int): # Handle other types like float if necessary
                    try:
                        mode_val = int(mode_val)
                    except (ValueError, TypeError):
                        print(f"Warning: Could not convert mode value '{mode_val}' for parent '{p}' to int. Using 0 as fallback.")
                        mode_val = 0
                baseline_evidence[p] = mode_val
            else:
                print(f"Error: Parent '{p}' (identified by model) not found in df_subset columns. Cannot calculate baseline.")
                all_parents_found_in_data = False
                # Decide how to handle: stop, or continue without this parent?
                # For now, we'll allow continuing but the prepare_evidence function will warn.

        if not all_parents_found_in_data:
             print("Warning: Not all parents found in data. Baseline evidence may be incomplete.")
        # else:
        #      print("Baseline Evidence (Mode of Parents):", baseline_evidence) # Optional print

    except Exception as e:
        print(f"Error getting parents or calculating baseline evidence: {e}")
        # Reset variables to indicate failure
        parents = []
        triage_parents_set = set()
        baseline_evidence = {}
else:
    print(f"\nCannot get parents or baseline: Model not loaded, data not loaded, or '{target_variable}' not in model.")


# --- Define the prepare_evidence function ---
def prepare_evidence(evidence_dict, all_parents_set, base_evidence):
    """
    Prepares an evidence dictionary for inference.
    Ensures evidence contains exactly all parent nodes of the target variable,
    with correct integer types, using baseline values for missing parents.
    """
    if not all_parents_set: # If parents couldn't be determined or were missing
        print("Warning: Cannot prepare evidence, parent set is empty or incomplete.")
        return None

    prepared_evidence = {}
    # Start with base evidence for all required parents
    for p in all_parents_set:
         if p in base_evidence:
             prepared_evidence[p] = base_evidence[p] # Already correct type (int)
         else:
             # This case happens if a parent was identified but missing from df_subset
             print(f"Warning: Parent '{p}' missing from baseline_evidence during preparation. Using 0 as fallback.")
             # Decide how to handle: skip, error, or use a default? Using 0 for now.
             prepared_evidence[p] = 0


    # Override with values from the specific evidence dict, but only if it's a parent
    for key, value in evidence_dict.items():
        if key in all_parents_set:
            # Ensure correct type (standard Python int)
            try:
                if isinstance(value, np.integer):
                    prepared_evidence[key] = int(value)
                elif isinstance(value, (int, float)): # Allow float input, cast to int
                     prepared_evidence[key] = int(value)
                else:
                     # Attempt conversion if it's string-like number? Risky.
                     prepared_evidence[key] = int(value)
            except (ValueError, TypeError):
                 print(f"Warning: Could not convert value '{value}' for key '{key}' to int. Using baseline value '{base_evidence.get(key, 0)}'.")
                 prepared_evidence[key] = base_evidence.get(key, 0) # Fallback to baseline or 0 if baseline also missing
        # else: # Optional: Warn about extra keys provided in the scenario dict
            # print(f"Note: Key '{key}' from evidence_dict is not a parent of {target_variable} and will be ignored.")


    # Final check: ensure all required parents are still in the dict
    # This shouldn't happen with the current logic but is a safeguard
    for p in all_parents_set:
        if p not in prepared_evidence:
             print(f"Error: Parent '{p}' was lost during evidence preparation. This should not happen.")
             return None # Indicate error

    return prepared_evidence

print(f"\nprepare_evidence function defined.")
if not triage_parents_set:
    print(f"Warning: Parent set for {target_variable} is empty, evidence preparation might fail.")
```

## 5. Execute Inference Queries
Use the `prepare_evidence` function to format each scenario's dictionary. Then, call `infer.query()` specifying `triage_level` as the target variable and passing the prepared evidence. Print the resulting probability distribution for `triage_level` for each scenario.

```python
# --- Prepare the specific evidence dictionaries ---
# Only proceed if the parent set and baseline were determined
if triage_parents_set and baseline_evidence:
    print("\n--- Preparing Evidence for Scenarios ---")
    low_evidence_prepared = prepare_evidence(low_evidence, triage_parents_set, baseline_evidence)
    moderate_evidence_prepared = prepare_evidence(moderate_evidence, triage_parents_set, baseline_evidence)
    high_evidence_prepared = prepare_evidence(high_evidence, triage_parents_set, baseline_evidence)
    cardiac_evidence_prepared = prepare_evidence(cardiac_evidence, triage_parents_set, baseline_evidence)
else:
    print("\n--- Skipping Evidence Preparation (Parents/Baseline unavailable) ---")
    low_evidence_prepared = None
    moderate_evidence_prepared = None
    high_evidence_prepared = None
    cardiac_evidence_prepared = None


# --- Perform Inference with Prepared Evidence ---
print("\n--- Inference with Prepared Evidence ---")

if infer is None:
    print("Inference engine not available. Skipping queries.")
else:
    # Query for Low Evidence
    print("\nLow Evidence (Prepared):", low_evidence_prepared)
    if low_evidence_prepared:
        try:
            result_low = infer.query(variables=[target_variable], evidence=low_evidence_prepared, show_progress=False)
            print(f"Result (Low Evidence) for {target_variable}:")
            print(result_low)
        except Exception as e:
            print(f"Error during low evidence query: {e}")
            # print("State names map:", state_names_map) # Debugging state names
            # print("Evidence values:", low_evidence_prepared) # Debugging evidence
    else:
        print("Skipping low evidence query due to preparation error or missing parents/baseline.")

    # Query for Moderate Evidence
    print("\nModerate Evidence (Prepared):", moderate_evidence_prepared)
    if moderate_evidence_prepared:
        try:
            result_moderate = infer.query(variables=[target_variable], evidence=moderate_evidence_prepared, show_progress=False)
            print(f"Result (Moderate Evidence) for {target_variable}:")
            print(result_moderate)
        except Exception as e:
            print(f"Error during moderate evidence query: {e}")
    else:
        print("Skipping moderate evidence query due to preparation error or missing parents/baseline.")

    # Query for High Evidence
    print("\nHigh Evidence (Prepared):", high_evidence_prepared)
    if high_evidence_prepared:
        try:
            result_high = infer.query(variables=[target_variable], evidence=high_evidence_prepared, show_progress=False)
            print(f"Result (High Evidence) for {target_variable}:")
            print(result_high)
        except Exception as e:
            print(f"Error during high evidence query: {e}")
    else:
        print("Skipping high evidence query due to preparation error or missing parents/baseline.")

    # Query for Cardiac Evidence
    print("\nCardiac Evidence (Prepared):", cardiac_evidence_prepared)
    if cardiac_evidence_prepared:
        try:
            result_cardiac = infer.query(variables=[target_variable], evidence=cardiac_evidence_prepared, show_progress=False)
            print(f"Result (Cardiac Evidence) for {target_variable}:")
            print(result_cardiac)
        except Exception as e:
            print(f"Error during cardiac evidence query: {e}")
    else:
        print("Skipping cardiac evidence query due to preparation error or missing parents/baseline.")
```

## 6. Analyze Feature Impact via Inference (Sensitivity Analysis)
Investigate how changing the state of specific parent variables affects the probability distribution of `triage_level`, while keeping other parents at their baseline (mode) values. This helps understand the influence of individual factors according to the learned model.

We analyze the impact of:
*   `has_myocardial_infarction`
*   `has_cardiac_arrest`
*   `condition_context`

```python
print("\n--- Analyzing Feature Impact on Triage Level ---")

if infer is None or not baseline_evidence or not triage_parents_set:
    print("Cannot perform impact analysis: Inference engine, baseline evidence, or parent set not available.")
else:
    # Variables to analyze (should be parents of triage_level)
    vars_to_analyze = ["has_myocardial_infarction", "has_cardiac_arrest", "condition_context"]

    # Get the CPD for triage_level to confirm parents if needed (optional check)
    triage_cpd = model.get_cpds(target_variable)
    if not triage_cpd:
        print(f"Warning: Could not retrieve CPD for {target_variable}. Parent check skipped.")
        model_parents = list(triage_parents_set) # Use the set derived earlier
    else:
        model_parents = triage_cpd.variables[1:] # Parents from CPD evidence list

    print(f"\nBaseline evidence (modes of parents) used for analysis: {baseline_evidence}")

    for var in vars_to_analyze:
        print(f"\n>>> Analyzing Impact of '{var}' on '{target_variable}':")

        if var not in triage_parents_set:
            print(f"   '{var}' is not in the set of direct parents ({triage_parents_set}). Skipping analysis for this variable.")
            continue
        if var not in baseline_evidence:
             print(f"   '{var}' is a parent but missing from baseline evidence. Skipping analysis.")
             continue

        # Get the possible states for this variable from the state_names_map
        possible_states = state_names_map.get(var)
        if not possible_states:
            print(f"   Warning: Could not find state names for '{var}' in state_names_map. Assuming binary [0, 1].")
            possible_states = [0, 1] # Fallback assumption

        print(f"   Possible states for '{var}': {possible_states}")

        # Perform inference for each state of the variable, keeping others at baseline
        for state in possible_states:
            # Create evidence by copying baseline and setting the variable's state
            evidence_copy = baseline_evidence.copy()
            evidence_copy[var] = state # Set the state for the variable under analysis

            # Re-prepare the evidence to ensure it's clean (might be redundant if baseline_evidence is already perfect)
            # Using the full triage_parents_set ensures all necessary parents are included.
            prepared_impact_evidence = prepare_evidence(evidence_copy, triage_parents_set, baseline_evidence)

            if prepared_impact_evidence:
                try:
                    # Ensure the state we are testing is valid according to the model's CPD state names for this variable
                    # This check is implicitly handled by pgmpy if state_names were set correctly during model fitting/loading.
                    # We rely on prepare_evidence to have handled type conversion.
                    
                    result_impact = infer.query(variables=[target_variable], evidence=prepared_impact_evidence, show_progress=False)
                    # Format output nicely
                    prob_dist = result_impact.values.tolist()
                    state_names_target = state_names_map.get(target_variable, [f"State_{i}" for i in range(len(prob_dist))])
                    prob_str = ", ".join([f"{state_names_target[i]}: {p:.4f}" for i, p in enumerate(prob_dist)])
                    print(f"   Inference with {var}={state} (others at baseline): P({target_variable}) = [{prob_str}]")

                except ValueError as ve:
                     print(f"   Error during inference for {var}={state}: {ve}")
                     print(f"   Problematic evidence: {prepared_impact_evidence}")
                     # Check if the state is actually present in the model's definition for this variable
                     if var in state_names_map and state not in state_names_map[var]:
                         print(f"   Potential issue: State '{state}' for variable '{var}' might not be defined in the model's state_names ({state_names_map[var]}).")
                except Exception as e:
                    print(f"   Error during inference for {var}={state}: {e}")
                    print(f"   Evidence used: {prepared_impact_evidence}")
            else:
                print(f"   Skipping inference for {var}={state} due to evidence preparation error.")

print("\n--- End of Analysis ---")
```
