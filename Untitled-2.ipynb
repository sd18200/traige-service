# Bayesian Network Inference

A Jupyter notebook demonstrating inference queries with a pre-trained Bayesian Network.

## Import Libraries and Load/Prepare Model for Inference

Import necessary libraries like `pgmpy.inference.VariableElimination` and load the pre-trained Bayesian Network model (e.g., from the exported JSON or directly if trained in the same notebook). Instantiate the inference engine (`VariableElimination`).

```python
import json
import numpy as np
import pandas as pd
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.inference import VariableElimination
from pgmpy.factors.discrete import TabularCPD

# --- Load Model from JSON ---
model_file = "learned_bn_model.json"
print(f"Loading model from {model_file}...")

try:
    with open(model_file, "r") as f:
        loaded_data = json.load(f)
except FileNotFoundError:
    print(f"Error: Model file '{model_file}' not found. Please ensure the model was trained and exported.")
    # Exit or raise error if the model file is essential
    raise SystemExit(f"Model file '{model_file}' not found.")
except json.JSONDecodeError:
    print(f"Error: Could not decode JSON from '{model_file}'. File might be corrupted.")
    raise SystemExit(f"Could not decode JSON from '{model_file}'.")


# --- Reconstruct the Model ---
model = DiscreteBayesianNetwork()

# Add nodes
model.add_nodes_from(loaded_data['nodes'])

# Add edges
model.add_edges_from([(str(u), str(v)) for u, v in loaded_data['edges']])

# Reconstruct and add CPDs
print("Reconstructing CPDs...")
reconstructed_cpds = []
for cpd_data in loaded_data['cpds']:
    variable = str(cpd_data['variable'])
    state_names = {str(var): [str(s) for s in states] for var, states in cpd_data['state_names'].items()}
    variable_card = len(state_names[variable])
    values = np.array(cpd_data['values']) # Should already be list of lists

    evidence = [str(ev) for ev in cpd_data['evidence']]
    evidence_card = [len(state_names[ev]) for ev in evidence]

    # Reshape values if necessary (pgmpy expects shape (var_card, ev1_card, ev2_card, ...))
    # The exported list might be flat or already correctly shaped depending on export.
    # Assuming it needs reshaping based on cardinalities.
    expected_shape = tuple([variable_card] + evidence_card)
    if values.size != np.prod(expected_shape):
         print(f"Warning: CPD values size mismatch for {variable}. Expected {np.prod(expected_shape)}, got {values.size}. Trying Fortran order reshape.")
         # Try reshaping assuming Fortran order if sizes match
         try:
             # pgmpy uses Fortran order for CPD value arrays
             values = values.reshape(expected_shape, order='F')
         except ValueError as e:
              print(f"Error reshaping CPD for {variable}: {e}. Skipping this CPD.")
              continue # Skip this CPD if reshape fails
    else:
         values = values.reshape(expected_shape, order='F')


    cpd = TabularCPD(
        variable=variable,
        variable_card=variable_card,
        values=values,
        evidence=evidence,
        evidence_card=evidence_card,
        state_names=state_names
    )
    reconstructed_cpds.append(cpd)

model.add_cpds(*reconstructed_cpds)

# --- Check Model Validity ---
try:
    model.check_model()
    print("✅ Model loaded and validated successfully.")
except Exception as e:
    print(f"Error validating the loaded model: {e}")
    # Decide how to proceed if model check fails
    raise SystemExit("Model validation failed.")


# --- Instantiate Inference Engine ---
infer = VariableElimination(model)
print("✅ Inference engine initialized.")


# --- Placeholder for Data Loading (Needed for Baseline Evidence) ---
# The calculation of baseline_evidence in the next step requires the original
# dataframe subset (df_subset) to calculate modes.
# Load it here if you need dynamic baseline calculation.
# Otherwise, you might define a static baseline_evidence dictionary.
print("\nNOTE: Loading data for baseline calculation (optional)...")
try:
    # Attempt to load the data used for training
    df_subset = pd.read_csv("Discretized_Dataset_with_Context.csv") # Load the original data
    # Apply necessary preprocessing steps to get df_subset consistent with training
    # This is a simplified version - ensure it matches the original preprocessing exactly
    ordinal_mapping = {'low': 0, 'medium': 1, 'high': 2}
    ordinal_columns = [
        'Heart rate', 'Systolic Blood Pressure', 'Diastolic Blood Pressure',
        'Respiratory rate', 'Body temperature',
        'Pain severity - 0-10 verbal numeric rating [Score] - Reported', 'age'
    ]
    for col in ordinal_columns:
        if col in df_subset.columns:
            df_subset[col] = df_subset[col].map(lambda x: ordinal_mapping.get(x, x))

    from sklearn.preprocessing import LabelEncoder
    label_encoders = {}
    for col in df_subset.select_dtypes(include=['object']).columns:
         if col in df_subset.columns:
             le = LabelEncoder()
             df_subset[col] = le.fit_transform(df_subset[col])
             label_encoders[col] = le # Store if needed, not used here

    selected_columns = loaded_data['nodes'] # Use nodes from the loaded model
    df_subset = df_subset[[col for col in selected_columns if col in df_subset.columns]]

    # Convert types to Int64 (matching original script)
    for col in df_subset.columns:
        if not pd.api.types.is_numeric_dtype(df_subset[col]):
            try:
                df_subset[col] = pd.to_numeric(df_subset[col], errors='coerce').astype('Int64')
            except Exception as e:
                 print(f"Warning: Could not convert column {col} to numeric during reload. Error: {e}")
        elif pd.api.types.is_float_dtype(df_subset[col]):
             df_subset[col] = df_subset[col].astype(float).astype('Int64')
        elif pd.api.types.is_integer_dtype(df_subset[col]):
             df_subset[col] = df_subset[col].astype('Int64')

    df_subset = df_subset.dropna()
    print("✅ Data loaded and preprocessed for baseline calculation.")

except FileNotFoundError:
    print("Warning: Dataset file not found. Baseline evidence calculation will fail if dynamic.")
    df_subset = None # Set to None if file not found
except Exception as e:
    print(f"Warning: Error loading or preprocessing data: {e}. Baseline calculation might fail.")
    df_subset = None
```

## Prepare Evidence for Querying

Define and use a helper function (`prepare_evidence`) to process the raw evidence dictionaries. This function should ensure the evidence passed to the query only contains direct parents of the target variable ('triage_level'), uses the correct data types (Python integers), and potentially handles missing parents by using baseline values (e.g., mode).

```python
# --- Get Parents and Baseline ---
target_variable = 'triage_level'
baseline_evidence = {}
triage_parents_set = set()

if target_variable not in model.nodes():
    print(f"Error: Target variable '{target_variable}' not found in the loaded model.")
else:
    try:
        parents = model.get_parents(target_variable)
        print(f"Parents of {target_variable}: {parents}")
        triage_parents_set = set(parents)

        # Calculate baseline evidence (mode of each parent) if df_subset is loaded
        if df_subset is not None and not df_subset.empty:
            all_parents_found_in_data = True
            for p in parents:
                if p in df_subset.columns:
                    # Calculate mode, ensure it's a standard Python int
                    mode_val = df_subset[p].mode()[0]
                    if isinstance(mode_val, np.integer):
                        mode_val = int(mode_val)
                    elif not isinstance(mode_val, int):
                         try:
                             mode_val = int(mode_val)
                         except (ValueError, TypeError):
                             print(f"Warning: Could not convert mode value '{mode_val}' for parent '{p}' to int. Using 0 as fallback.")
                             mode_val = 0
                    baseline_evidence[p] = mode_val
                else:
                    print(f"Error: Parent '{p}' not found in df_subset columns for baseline calculation.")
                    all_parents_found_in_data = False
                    # break # Option: Stop if a parent is missing in data

            if not all_parents_found_in_data:
                 print("Warning: Not all parents found in data. Baseline evidence might be incomplete.")
                 # Decide fallback: Use empty baseline, raise error, or use defaults?
                 # Using potentially incomplete baseline for now.
            print("Baseline Evidence (Modes):", baseline_evidence)
        else:
            print("Warning: df_subset not available. Cannot calculate baseline evidence from data.")
            # Option: Define a static baseline dictionary here if needed
            # baseline_evidence = {'parent1': 0, 'parent2': 1, ...} # Example static baseline

    except Exception as e:
        print(f"Error getting parents or calculating baseline evidence: {e}")
        # Reset parents and baseline if error occurs
        parents = []
        triage_parents_set = set()
        baseline_evidence = {}


# --- Define the prepare_evidence function ---
def prepare_evidence(evidence_dict, all_parents_set, base_evidence):
    """Ensures evidence has all parents, correct types, and no extra keys."""
    if not all_parents_set: # If parents couldn't be determined or target var missing
        print("Warning: Cannot prepare evidence, parent set is empty or target variable invalid.")
        return None

    prepared_evidence = {}
    # Start with base evidence for all parents (if available)
    for p in all_parents_set:
         if p in base_evidence:
             prepared_evidence[p] = base_evidence[p] # Already correct type from calculation
         else:
             # If parent not in baseline (e.g., data missing, static baseline not defined)
             # Use a default value, like 0, or raise an error. Using 0 for now.
             print(f"Warning: Parent '{p}' missing from baseline_evidence. Using default value 0.")
             prepared_evidence[p] = 0


    # Override with values from the specific evidence dict if it's a parent
    for key, value in evidence_dict.items():
        if key in all_parents_set:
            # Ensure correct type (standard Python int)
            try:
                if isinstance(value, np.integer):
                    prepared_evidence[key] = int(value)
                elif isinstance(value, (int, float)):
                     prepared_evidence[key] = int(value) # Cast float to int
                else:
                     # Attempt conversion if it's string-like number? Risky.
                     prepared_evidence[key] = int(value)
            except (ValueError, TypeError):
                 print(f"Warning: Could not convert value '{value}' for key '{key}' to int. Using baseline/default value '{prepared_evidence[key]}'.")
                 # Value remains the baseline/default already set

    # Final check: ensure all parents are still in the dict (should be guaranteed by starting with all parents)
    # for p in all_parents_set:
    #     if p not in prepared_evidence:
    #          print(f"Error: Parent {p} was lost during evidence preparation.")
    #          return None # Indicate error

    return prepared_evidence

# --- Prepare the specific evidence dictionaries ---
print("\nPreparing evidence dictionaries...")
low_evidence_prepared = prepare_evidence(low_evidence, triage_parents_set, baseline_evidence)
moderate_evidence_prepared = prepare_evidence(moderate_evidence, triage_parents_set, baseline_evidence)
high_evidence_prepared = prepare_evidence(high_evidence, triage_parents_set, baseline_evidence)
cardiac_evidence_prepared = prepare_evidence(cardiac_evidence, triage_parents_set, baseline_evidence)

print("✅ Evidence preparation complete.")
```

## Perform Inference Queries

Use the `infer.query()` method from the `VariableElimination` object. Pass the target variable(s) (e.g., `variables=['triage_level']`) and the prepared evidence dictionary for each scenario (e.g., `evidence=low_evidence_prepared`). Print the resulting probability distributions for the target variable under each evidence scenario.

```python
# --- Perform Inference with Prepared Evidence ---
print("\n--- Inference with Prepared Evidence ---")

target_variable = 'triage_level' # Ensure this matches the variable name in the model

# Function to run and print query results safely
def run_query(evidence_name, prepared_evidence):
    print(f"\n{evidence_name} (Prepared Evidence): {prepared_evidence}")
    if prepared_evidence is not None and target_variable in model.nodes():
        try:
            # Ensure evidence values are valid states for the respective variables
            # (pgmpy >= 0.1.18 performs checks)
            result = infer.query(variables=[target_variable], evidence=prepared_evidence, show_progress=False)
            print(f"Result ({evidence_name}):")
            print(result)
        except ValueError as ve:
             print(f"Error during {evidence_name} query: {ve}")
             print("This might indicate an invalid state value in the evidence for a variable.")
        except Exception as e:
            print(f"Error during {evidence_name} query: {e}")
            # You might want to print traceback for debugging
            # import traceback
            # traceback.print_exc()
    elif prepared_evidence is None:
        print(f"Skipping {evidence_name} query due to preparation error.")
    else:
         print(f"Skipping {evidence_name} query because target variable '{target_variable}' is not in the model.")

# Run queries for each scenario
run_query("Low Evidence", low_evidence_prepared)
run_query("Moderate Evidence", moderate_evidence_prepared)
run_query("High Evidence", high_evidence_prepared)
run_query("Cardiac Evidence", cardiac_evidence_prepared)

print("\n--- Inference Complete ---")
```
