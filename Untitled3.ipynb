```python
import pandas as pd

df = pd.read_csv("Modeling_Dataset_with_Encoded_Triage_and_Conditions.csv")

# Define numeric features (not to include)
numerical = [
    'Unnamed: 0', 'age', 'Body Height', 'Body Mass Index', 'Body Weight',
    'Body temperature', 'Diastolic Blood Pressure', 'Heart rate',
    'Pain severity - 0-10 verbal numeric rating [Score] - Reported',
    'Respiratory rate', 'Systolic Blood Pressure', 'triage_level', 'triage_reason'
]

# All 0/1 binary columns excluding numerical
binary = [col for col in df.columns
          if set(df[col].dropna().unique()).issubset({0, 1}) and col not in numerical]

# Compare against current known_conditions
with open("known_conditions.txt") as f:
    known = [line.strip() for line in f]

# Show missing ones
missing = list(set(binary) - set(known))
print(f"🧩 Missing columns: {missing}")
print(f"➡️ Add these {len(missing)} to known_conditions.txt")
```
