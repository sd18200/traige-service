```python
import pandas as pd

# Load dataset
df = pd.read_csv("final_patient_dataset_with_vital_signs_only.csv")

# Convert key vital sign columns to numeric
cols_to_num = [
    "Heart rate", "Respiratory rate", "Systolic Blood Pressure", 
    "Diastolic Blood Pressure", "Pain severity - 0-10 verbal numeric rating [Score] - Reported",
    "Hemoglobin [Mass/volume] in Blood", 
    "Platelets [#/volume] in Blood by Automated count"
]
for col in cols_to_num:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors="coerce")

# Clean condition keywords if available
if "DESCRIPTION" in df.columns:
    df["DESCRIPTION"] = df["DESCRIPTION"].fillna("").str.lower()
elif "conditions" in df.columns:
    df["conditions"] = df["conditions"].fillna("").str.lower()
else:
    df["conditions"] = ""

def triage_score_with_reason(row):
    reasons = []
    hr = row.get("Heart rate")
    rr = row.get("Respiratory rate")
    sys = row.get("Systolic Blood Pressure")
    dia = row.get("Diastolic Blood Pressure")
    pain = row.get("Pain severity - 0-10 verbal numeric rating [Score] - Reported")
    hb = row.get("Hemoglobin [Mass/volume] in Blood")
    plt = row.get("Platelets [#/volume] in Blood by Automated count")
    cond = row.get("conditions", "")
    enc = row.get("ENCOUNTERCLASS", "").lower()

    # ----- URGENT -----
    if (
        (pd.notna(hr) and hr > 130) or
        (pd.notna(rr) and rr > 30) or
        (pd.notna(sys) and sys < 100) or
        (pd.notna(dia) and dia > 110) or
        (pd.notna(hb) and hb < 8) or
        (pd.notna(plt) and plt < 50) or
        (pd.notna(pain) and pain >= 9) or
        any(x in cond for x in ["myocardial infarction", "stroke", "sepsis", "pulmonary embolism", "cardiac arrest"])
    ):
        if pd.notna(hr) and hr > 130: reasons.append(f"HR {hr} bpm (tachycardia)")
        if pd.notna(rr) and rr > 30: reasons.append(f"RR {rr} bpm (resp distress)")
        if pd.notna(sys) and sys < 100: reasons.append(f"SBP {sys} mmHg (shock)")
        if pd.notna(dia) and dia > 110: reasons.append(f"DBP {dia} mmHg (HTN crisis)")
        if pd.notna(hb) and hb < 8: reasons.append(f"Hemoglobin {hb} g/dL (anemia crisis)")
        if pd.notna(plt) and plt < 50: reasons.append(f"Platelets {plt}k/uL (bleeding risk)")
        if pd.notna(pain) and pain >= 9: reasons.append(f"Pain score {pain} (severe)")
        for kw in ["myocardial infarction", "stroke", "sepsis", "pulmonary embolism", "cardiac arrest"]:
            if kw in cond:
                reasons.append(f"Condition: {kw}")
        return pd.Series(["urgent", "; ".join(reasons)])

    # ----- MODERATE -----
    if (
        (pd.notna(hr) and 101 <= hr <= 130) or
        (pd.notna(rr) and 21 <= rr <= 30) or
        (pd.notna(sys) and 100 <= sys <= 110) or
        (pd.notna(hb) and 8 <= hb <= 11) or
        (pd.notna(plt) and 50 <= plt <= 150) or
        (pd.notna(pain) and 5 <= pain <= 8) or
        any(x in cond for x in ["pneumonia", "asthma", "flu", "hypertension", "copd"])
    ):
        if pd.notna(hr) and hr > 100: reasons.append(f"HR {hr} bpm")
        if pd.notna(rr) and rr > 20: reasons.append(f"RR {rr} bpm")
        if pd.notna(sys) and sys <= 110: reasons.append(f"SBP {sys} mmHg")
        if pd.notna(hb) and hb <= 11: reasons.append(f"Hemoglobin {hb} g/dL")
        if pd.notna(plt) and plt <= 150: reasons.append(f"Platelets {plt}k/uL")
        if pd.notna(pain) and pain >= 5: reasons.append(f"Pain score {pain}")
        for kw in ["pneumonia", "asthma", "flu", "hypertension", "copd"]:
            if kw in cond:
                reasons.append(f"Condition: {kw}")
        return pd.Series(["moderate", "; ".join(reasons)])

    # ----- LOW -----
    reasons.append("Vitals and symptoms within normal limits")
    return pd.Series(["low", "; ".join(reasons)])

# Apply scoring
df[["triage_level", "triage_reason"]] = df.apply(triage_score_with_reason, axis=1)

# Save result
df.to_csv("enhanced_realistic_triage_with_reasons.csv", index=False)
print("Saved: enhanced_realistic_triage_with_reasons.csv")
```
