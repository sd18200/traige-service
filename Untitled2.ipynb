```python
import pandas as pd
from sklearn.preprocessing import MultiLabelBinarizer

# === Load your dataset ===
df = pd.read_csv("Cleaned_Dataset_for_Modeling.csv")

# === Step 1: Encode triage_level (target variable) ===
# Map 'low' → 0 (P3), 'moderate' → 1 (P2), 'urgent' → 2 (P1)
triage_mapping = {"low": 0, "moderate": 1, "urgent": 2}
df["triage_level"] = df["triage_level"].map(triage_mapping)

# === Step 2: Parse 'conditions' column into lists ===
conditions_lists = df['conditions'].fillna('').apply(
    lambda x: [c.strip().lower() for c in x.split(',') if c.strip()]
)

# === Step 3: Initialize MultiLabelBinarizer and encode ===
mlb = MultiLabelBinarizer(sparse_output=False)
encoded_array = mlb.fit_transform(conditions_lists)

# === Step 4: Convert to DataFrame with clean column names ===
encoded_conditions_df = pd.DataFrame(
    encoded_array,
    columns=[f"has_{c.replace(' ', '_').replace('(', '').replace(')', '')}" for c in mlb.classes_]
)
encoded_conditions_df.index = df.index

# === Step 5: Combine encoded condition features with original DataFrame ===
df_combined = pd.concat([df.drop(columns=["conditions"]), encoded_conditions_df], axis=1)

# === Step 6: Save final dataset ===
df_combined.to_csv("Modeling_Dataset_with_Encoded_Triage_and_Conditions.csv", index=False)
print(" Dataset saved as: Modeling_Dataset_with_Encoded_Triage_and_Conditions.csv")
```
